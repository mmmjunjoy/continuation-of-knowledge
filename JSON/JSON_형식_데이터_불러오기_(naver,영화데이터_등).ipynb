```python
# 모듈 설치


import json
import urllib.request
import pprint

# 여기서 pprint는 결과를 보기 좋게 출력해주는 기능이다 , json 과는 상관 없다.
```

```python
# naver main 화면 데이터 불러오기

url = 'https://www.naver.com'
jsonurl = urllib.request.urlopen(url).read().decode('utf-8')
#byte_data = jsonurl.read()
#text_data = byte_data.decode('utf-8')

jsonurl
```

```python
# json 형식의 데이터 불러오기

url = 'https://python.bakyeono.net/data/movies.json'
text_data = urllib.request.urlopen(url).read().decode('utf-8')
movies = json.loads(text_data)

movies
```
