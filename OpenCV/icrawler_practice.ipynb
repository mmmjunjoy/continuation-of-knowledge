```python
from icrawler.builtin import  GoogleImageCrawler
```

```python
google_crawler = GoogleImageCrawler(
    feeder_threads =1,
    parser_threads=1,
    downloader_threads =4,
    storage ={'root_dir': '/content/ laughing person'})
filters =dict(
    size='large',
    color='orange',
    license='commercial,modify',
    date=((2018,1,1),(2022,11,30))
)

a = input()

google_crawler.crawl(keyword=a,filters=filters,offset=0,max_num=1000,
                     min_size=(200,200),max_size=None,file_idx_offset=0)

```
