```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
face_cascasde = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
image = cv2.imread('/content/junjoyoung.jpg')
```

```python
image = cv2.resize(image, (250, 250))
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(grayImage,cmap='gray')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
faces = face_cascasde.detectMultiScale(grayImage,1.03,5)

print(faces.shape)
print("Number of faces detected:" + str(faces.shape[0]))
```

```python
faces
```

```python
for(x,y,w,h) in faces: 
  # if(w+h > 100):
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)

print(image.shape[0])
print(image.shape[1])
# cv2.rectangle(image,((0,image.shape[0] )),
#               (1,image.shape[1]),(255,0,0),2);

# cv2.putText(image, "PinkWink test", (0,image.shape[0] -10), 
#             cv2.FONT_HERSHEY_TRIPLEX,0.5,(0,0,0),-1);

plt.figure(figsize=(12,12))
plt.imshow(image)
plt.xticks([]),plt.yticks([])
plt.show()
```
